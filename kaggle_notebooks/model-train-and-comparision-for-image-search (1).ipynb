🟩 CELL 1: Install & Dataset Prep 🟩

```python
!pip install ultralytics

from ultralytics import YOLO
import torch
import os
import pandas as pd

print(f"GPU Available: {torch.cuda.is_available()}")

source_dir = "/kaggle/input"
root_data_dir = None
for root, dirs, files in os.walk(source_dir):
    if 'train' in dirs:
        root_data_dir = root
        break

working_dataset = '/kaggle/working/car_parts_dataset_linked'
os.makedirs(working_dataset, exist_ok=True)

for split in ['train', 'valid', 'test']:
    src = os.path.join(root_data_dir, split)
    if os.path.exists(src):
        link_name = 'val' if split == 'valid' else split
        dst = os.path.join(working_dataset, link_name)
        if not os.path.exists(dst):
            os.symlink(src, dst)
print("Dataset prep complete!")
```

🟩 CELL 2: Train MEDIUM Model 🟩

```python
from ultralytics import YOLO

print("🚀 TRAINING YOLO11 MEDIUM...")
model_m = YOLO('yolo11m-cls.pt')

model_m.train(
    data='/kaggle/working/car_parts_dataset_linked',
    epochs=30,      
    imgsz=224,      
    device=0,       
    project='/kaggle/working/cloud_models',
    name='yolo11_medium_run'
)
print("✅ MEDIUM Training Complete and Saved!")
```

🟩 CELL 3: Train LARGE Model 🟩

```python
from ultralytics import YOLO

print("🚀 TRAINING YOLO11 LARGE...")
model_l = YOLO('yolo11l-cls.pt')

model_l.train(
    data='/kaggle/working/car_parts_dataset_linked',
    epochs=30,      
    imgsz=224,      
    device=0,       
    project='/kaggle/working/cloud_models',
    name='yolo11_large_run'
)
print("✅ LARGE Training Complete and Saved!")
```

🟩 CELL 4: Train EXTRA LARGE Model 🟩

```python
from ultralytics import YOLO

print("🚀 TRAINING YOLO11 EXTRA LARGE...")
model_x = YOLO('yolo11x-cls.pt')

model_x.train(
    data='/kaggle/working/car_parts_dataset_linked',
    epochs=30,      
    imgsz=224,      
    device=0,       
    project='/kaggle/working/cloud_models',
    name='yolo11_extralarge_run'
)
print("✅ EXTRA LARGE Training Complete and Saved!")
```

🟩 CELL 5: Results & Comparison 🟩

```python
import pandas as pd
import os

models_trained = ['medium', 'large', 'extralarge']
results_dict = {}

for size in models_trained:
    run_folder = f"/kaggle/working/cloud_models/yolo11_{size}_run"
    csv_file = os.path.join(run_folder, "results.csv")
    
    if os.path.exists(csv_file):
        df = pd.read_csv(csv_file)
        df.columns = df.columns.str.strip()
        
        best_top1 = df['metrics/accuracy_top1'].max()
        best_top5 = df['metrics/accuracy_top5'].max()
        
        results_dict[size.capitalize()] = {
            'Top-1 Accuracy': round(best_top1, 4),
            'Top-5 Accuracy': round(best_top5, 4)
        }

print("\n🏆 FINAL CLOUD MODEL ACCURACY COMPARISON 🏆")
if results_dict:
    df_results = pd.DataFrame(results_dict).T
    print(df_results[['Top-1 Accuracy', 'Top-5 Accuracy']])
else:
    print("Could not find results.csv files.")
```
